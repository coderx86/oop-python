### 🧠 Object-Oriented Programming (OOP)

Object-Oriented Programming (OOP) is a programming paradigm where we use **objects** (combination of data and behavior) to represent real-world entities. A **class** works as a blueprint, and an **object** is an instance of that class.

OOP is built on four main principles: 
- **Encapsulation** (hiding and protecting data) 
- **Inheritance** (reusing code from other classes)
- **Polymorphism** (same method behaving differently)
- **Abstraction** (hiding unnecessary details and showing only essential features).

It helps in writing **clean, reusable, and maintainable code**, especially for large and complex systems.

### 📦Class & Object

* **Class**: A blueprint/template to create objects (defines properties & behavior).
* **Object**: An instance of a class (real entity created from the class).

---

### 🔹 Step-by-Step Implementation

**1. Define a class**

```python
class Person:
    pass
```

**2. Add constructor (`__init__`)**

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

**3. Add methods (behavior)**

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        print(f"Hi, I am {self.name}")
```

**4. Create object**

```python
p1 = Person("John", 25)
```

**5. Access properties & methods**

```python
print(p1.name)     # Access attribute
p1.greet()         # Call method
```


````python
class Phone:
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price =price

    def showInfo(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Price: {self.price}")

# Create objects of the Phone class
phone1 = Phone("Apple", "iphone 17 pro max", 1200)
phone2 = Phone("Samsung", "Galaxy S23 Ultra", 1100)

# Use objects to call the showInfo method
phone1.showInfo()
phone2.showInfo()
````
